```python
import numpy as np
import pandas as pd

alpha='5'
beta='5'
displacement = pd.read_csv('displacement-alpha'+alpha+'-beta'+beta+'.csv', header=None)
rotation = pd.read_csv('rotation-alpha'+alpha+'-beta'+beta+'.csv', header=None)
moment = pd.read_csv('moment-alpha'+alpha+'-beta'+beta+'.csv', header=None)
curvature = pd.read_csv('curvature-alpha'+alpha+'-beta'+beta+'.csv', header=None)
LPF =  pd.read_csv('LPF-alpha'+alpha+'-beta'+beta+'.csv', header=None)
SK =  pd.read_csv('SK-alpha'+alpha+'-beta'+beta+'.csv', header=None)
```

```python
Numero_iteracoes=20
Passo=30

disp=[]
maxes=[]    
mins=[]
rot=[]
mom=[]
cur=[]
load=[]
sk=[]
for i in range(0,Numero_iteracoes):
    disp.append(displacement.iloc[i][:])
    rot.append(rotation.iloc[i][:])
    mom.append(moment.iloc[i][:])
    cur.append(curvature.iloc[i][:])
    load.append(int((i+1)*Passo*LPF.iloc[i][0]))
    sk.append(SK.iloc[i][:])

xx=rotation.iloc[-1][:]
intpoints=moment.iloc[-1][:]

xticks=[]
for i in range(0,121,4):
    xticks.append(xx[i])
    
    
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

def make_graph(_xx,_var):
    figure(num=None, figsize=(20, 12), dpi=80, facecolor='w', edgecolor='k')
    maxes=[]
    mins=[]
    for i in range(0,Numero_iteracoes):
        maxes.append(max(_var[i]))
        mins.append(min(_var[i]))
        plt.plot(_xx,_var[i],label=str(load[i])+'N') 

    stepd=(max(maxes)-min(mins))/20.0
    plt.yticks(np.arange(min(mins), max(maxes)+stepd ,step=stepd))
    plt.xticks(xticks)
    plt.grid(True)

    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
    plt.close()
```

```python
make_graph(xx,disp)
```

```python
make_graph(intpoints,mom)
```

```python
make_graph(xx,rot)
```

```python
make_graph(intpoints,sk)
```

```python
make_graph(intpoints,cur)
```
